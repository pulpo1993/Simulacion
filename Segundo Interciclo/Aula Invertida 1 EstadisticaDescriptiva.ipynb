#### 1. analizar la estadistica descriptiva los datos de la fifa

```python
datosFifa = pd.read_csv('Recursos/fifa_datos.csv',',')
datosFifa.head()
```

```python
#buscamos la correlacion que existe en este dataset con los datos que lo contienen
# Correlacion
datosFifa.corr()
```
